```python
a=np.random.randint(0,10,10)
b=np.random.randint(0,10,10)
c=np.random.randint(0,10,10)
```

```python
#scatter plot
plt.scatter(a,b,c='red',marker='o' , s=50,label='A,B')
plt.scatter(a,c,c='green',marker='s' , s=50,label='A,C')
plt.title("Scatter plot")
plt.xlabel("$X$ axis")
plt.ylabel("$Y$ axis")
plt.legend()
plt.show()
```

```python
#line plot
plt.plot(a,b,c='red',linestyle="--",marker='s',label='A,B')
plt.plot(a,c,c='green',marker='o',label='A,C')
plt.title("Line plot")
plt.xlabel("$X$ axis")
plt.ylabel("$Y$ axis")
plt.legend()
plt.show()
```

```python
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
data = load_boston()
```

```python
data
```

```python
plt.hist(data.target)
plt.xlabel('price')
plt.ylabel('count')
plt.show()
```

```python
X = data['data']
y = data['target']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
lr.score(X_train, y_train)
```

```python
lr.score(X_test,y_test)
```
